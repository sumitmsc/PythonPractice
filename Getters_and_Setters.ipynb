ðŸ§© What are Getters and Setters?
ðŸ”¹ Definition:

Getter â†’ A method that gets (returns) the value of a variable.

Setter â†’ A method that sets (updates) the value of a variable.

They help control access to class attributes â€” especially when you donâ€™t want variables to be directly modified from outside the class.

```python
#ðŸ§  Example Without Getter/Setter

class Person:
    def __init__(self, name):
        self.name = name

p = Person("Sumit")
print(p.name)       # Access directly
p.name = "Divya"    # Modify directly
print(p.name)

#Here, anyone can change name directly â€” even set it to something invalid like p.name = 123.

#To protect it, we can use getter and setter methods.
```

```python
#ðŸ§© Using Getter and Setter Methods (Traditional Way)
class Person:
    def __init__(self, name):
        self.__name = name   # note the double underscore â†’ private variable

    def get_name(self):      # Getter method
        return self.__name

    def set_name(self, value):  # Setter method
        if isinstance(value, str):
            self.__name = value
        else:
            print("Name must be a string!")

p = Person("Sumit")
print(p.get_name())    # uses getter

p.set_name("Divya")    # uses setter
print(p.get_name())

p.set_name(123)        # invalid update

#This ensures data validation â€” you control what can be set.
```

```python
# Pythonic Way: Using @property Decorator

#Python gives a cleaner and modern syntax for getters and setters using @property.

class Person:
    def __init__(self, name):
        self.__name = name

    @property
    def name(self):        # acts like getter
        return self.__name

    @name.setter
    def name(self, value): # acts like setter
        if isinstance(value, str):
            self.__name = value
        else:
            raise ValueError("Name must be a string!")

p = Person("Sumit")
print(p.name)      # getter is called

p.name = "Divya"   # setter is called
print(p.name)

p.name = 123       # raises ValueError
```
