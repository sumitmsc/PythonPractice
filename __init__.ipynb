```python
class Person:

  def __init__(self,n,o):
    self.name=n
    self.occ=o

  def info(self):
    print(f'My name is {self.name} and I am a {self.occ}')


a=Person('Harry','Developer')
b=Person('Divya','HR')

a.info()
b.info()



```

Explanation

class Person:
This line defines a class named Person.
A class is like a blueprint for creating objects (instances).

def __init__(self, n, o):
This is the constructor method, automatically called when a new object is created.

self refers to the current object.

n and o are parameters that represent name and occupation.

self.name = n
Stores the name value in the object’s name attribute.

self.occ = o
Stores the occupation value in the object’s occ attribute.

def info(self):
A method inside the class that prints a message using the object’s data.

Creating objects:
a = Person('Harry', 'Developer')
b = Person('Divya', 'HR')
a is a Person object with name "Harry" and occupation "Developer".

b is a Person object with name "Divya" and occupation "HR".

Calling methods:

a.info()
b.info()

Each call prints details of that specific object.
