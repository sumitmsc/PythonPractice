üß© What is a Decorator?

A decorator is a function that modifies or adds extra behavior to another function ‚Äî without changing the function‚Äôs actual code.

You can think of it like ‚Äúwrapping‚Äù a function to give it some extra power.

üéØ Real-life Example (Simple Analogy)

Imagine a simple coffee cup ‚òï.
You can add decorations to it ‚Äî like a lid or a sleeve ‚Äî without changing the coffee inside.

Similarly, in Python, a decorator adds something extra to a function.

```python
def decorator_function(func):
    def wrapper():
        print("Before the function runs...")
        func()  # call the original function
        print("After the function runs...")
    return wrapper

def say_hello():
    print("Hello!")

decorated_func = decorator_function(say_hello)
decorated_func()

```

```python
def decorator_fun(fun):
  def wrapper():
    print('Before function')
    fun()
    print('After function')
  return wrapper

def hello():
  print('Hello World!')

#hello()

decorator=decorator_fun(hello)

decorator()
```

```python
#Decorator
def greet(fx):
  def wrapper():
    print('Before function')
    fx()
    print('After function')
  return wrapper

@greet
def hello():
 print('Hello World!')

hello()
```

üß© Why Decorators Are Useful

You can use decorators for:

Logging

Authentication checks

Measuring execution time

Modifying input/output

Avoiding repetitive code

```python
#Example: Logging Decorator

def log_decorator(func):
    def wrapper():
        print(f"Function '{func.__name__}' is starting...")
        func()
        print(f"Function '{func.__name__}' finished.")
    return wrapper

@log_decorator
def test_api():
    print("Calling API...")

test_api()
```

Summary
Concept	Meaning
Decorator	A function that wraps another function to extend its behavior.
@decorator_name	Shortcut syntax to apply a decorator.
Use cases	Logging, timing, validation, authentication, etc.

```python
# Using *args and **kwargs

def decorator(fun):
  def wrapper(*args):
    print('Good morning!')
    fun(*args)
    print('Thanks for using function for addition')

  return wrapper

@decorator
def add(a,b):
  c=a+b
  print(f'addition of {a} and {b} is {c}')

add(5,7)
```
